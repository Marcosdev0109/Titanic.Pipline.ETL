```python
import pandas as pd
import requests
import json
from google import genai

# --- Configura√ß√£o da chave de API ---
# ‚ö†Ô∏è Recomendado: use vari√°vel de ambiente para seguran√ßa
# No terminal: export GOOGLE_API_KEY="sua_chave_aqui"
API_KEY = "AIzaSyC87jI_gJkFdbMDU1XnpQdU1kw29-s3acU"  # apenas para testes locais

# Cria o cliente Gemini
client = genai.Client(api_key=API_KEY)

# Lista de IDs de exemplo
user_ids = [1, 2, 3, 4, 5]

# Fun√ß√£o para buscar todos os usu√°rios
def get_all_users():
    response = requests.get('https://jsonplaceholder.typicode.com/users')
    return response.json() if response.status_code == 200 else None

# Fun√ß√£o que gera uma not√≠cia com base nos dados do usu√°rio
def gerador_de_noticias(user_data):
    prompt = (
        f"vo√ße √© o responsavel comercial de uma agencia de viagens que vende passagens para o Titanic, vo√ße esta entrando em contato com {user_data['name']}, decidiu viajar no titanic"
        f"que mora em {user_data['address']['city']} e trabalha na empresa {user_data['company']['name']}. para oferecer os produtos, conven√ßa os de modo criativo a embarcar nesta viagem"
        "Use sua habilidade e marketing para mostrar a estes potenciais clientes, o que eles poderao esperar nesta viagem, e em tom bem humorado passe o valor promocional, com at√© 2 frases."
    )

    # Gera conte√∫do com o Gemini 2.5 Flash
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
    )

    return response.text.strip()

# --- Execu√ß√£o principal ---
all_users = get_all_users()

if all_users is not None:
    # Filtra os usu√°rios desejados
    users = [u for u in all_users if u['id'] in user_ids]

    # Gera not√≠cias para cada usu√°rio
    for user in users:
        news = gerador_de_noticias(user)
        print(f"\nüì∞ Not√≠cia sobre {user['name']}:\n{news}\n")
else:
    print("Falha ao obter usu√°rios.")
```

```python
# --- Etapa de Load: Salvando as not√≠cias em um arquivo CSV ---

import csv

# Crie uma lista para armazenar os dados formatados para CSV
news_data_for_csv = []

# Adicione o cabe√ßalho do CSV
# Assumindo que voc√™ quer User ID, User Name e News
header = ['user_id', 'user_name', 'news']
news_data_for_csv.append(header)


# Simular a obten√ß√£o das not√≠cias geradas (como no exemplo JSON anterior)
# Em um pipeline real, voc√™ usaria a lista de not√≠cias geradas na etapa anterior
generated_news_list = []
if all_users is not None:
    filtered_users = [u for u in all_users if u['id'] in user_ids]
    for user in filtered_users:
         # Chame a fun√ß√£o novamente para obter a not√≠cia para este usu√°rio
         news = gerador_de_noticias(user) # Chame a fun√ß√£o novamente
         generated_news_list.append({
             "user_id": user['id'],
             "user_name": user['name'],
             "news": news
         })

# Agora, formate os dados para CSV e adicione √† lista
for item in generated_news_list:
    news_data_for_csv.append([item['user_id'], item['user_name'], item['news']])


# Define o nome do arquivo de sa√≠da
output_filename_csv = 'noticias_geradas.csv'

# Salva os dados em um arquivo CSV
with open(output_filename_csv, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(news_data_for_csv)

print(f"\n‚úÖ Not√≠cias salvas com sucesso em '{output_filename_csv}'")
```
